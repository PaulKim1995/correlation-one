```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import tools
import matplotlib.pyplot as plt
```

```python
data, amentities = tools.load_listings()
data['self_check_in'] = data['self_check-in']
data['family_friendly'] = data['family/kid_friendly']
data = data.dropna()
```

```python
model_params = 'review_scores_rating ~ review_scores_cleanliness + review_scores_communication + review_scores_location + review_scores_checkin + wireless_internet + air_conditioning + kitchen + heating + pets_allowed + cable_tv + family_friendly + cancellation_policy_flexible + instant_bookable'

```

# Subsetting based on metro area

```python
for name, group in data.groupby('metropolitan'):
    results = smf.ols(model_params, data=group).fit()
    
    print('metro_area', name)
    print(results.summary())
```

# Subsetting based on number of bedrooms

```python
for name, group in data.groupby('bedrooms'):
    results = smf.ols(model_params, data=group).fit()
    
    print('number of bedrooms:', name)
    print(results.summary())
    if name == 4:
        break
```

# Subsetting based on number of room type

```python
for name, group in data.groupby('room_type'):
    results = smf.ols(model_params, data=group).fit()
    
    print('room type:', name)
    print(results.summary())
```

# Overall

```python
results = smf.ols(model_params, data=data).fit()
print(results.summary())
```
